```python
import pdfplumber
import os
print(f" Directorio actual de trabajo: {os.getcwd()}")
```

```python
archivo_pdf = "../cv_prueba.pdf" 

if os.path.exists(archivo_pdf):
    print(" Archivo encontrado. ¡Listo para operar!")
else:
    print(f"No encuentro '{archivo_pdf}'. Revisa la ruta.")
```

```python
texto_completo = ""

with pdfplumber.open(archivo_pdf) as pdf:
    print(f"El PDF tiene {len(pdf.pages)} páginas.")
    
    # Vamos a analizar solo la primera página a fondo
    primera_pagina = pdf.pages[0]
    
    # 1. Extraer texto simple
    texto = primera_pagina.extract_text()
    print("\n---  VISTA PREVIA DEL TEXTO ---")
    print(texto[:300]) # Primeros 300 caracteres
    print("...")
    
    # 2. Verificar si tiene tablas
    tablas = primera_pagina.extract_tables()
    if tablas:
        print(f"\n---  TABLAS DETECTADAS ({len(tablas)}) ---")
        for i, tabla in enumerate(tablas):
            print(f"Tabla {i+1}: {tabla}")
    else:
        print("\n---  No se detectaron tablas estructuradas en la pág 1 ---")
```
